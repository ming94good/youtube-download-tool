```python
import os
```

```python
import subprocess

# define the ls command
ls = subprocess.Popen(["ls", "-p", "."],
                      stdout=subprocess.PIPE,
                     )

# define the grep command
grep = subprocess.Popen(["grep", "-v", "/$"],
                        stdin=ls.stdout,
                        stdout=subprocess.PIPE,
                        )

# read from the end of the pipe (stdout)
endOfPipe = grep.stdout

# output the files line by line
file_list = []
for line in endOfPipe:
    filename = line.decode()
    print (filename)
    filename_format = filename.split('\n')[0]
    file_list.append(filename_format)
```

```python
num = 36
for file in file_list:
    if num > 113:
        break
    os.rename(str(file),'../video/'+ str(num) +'.mp4')
    num += 1
```
